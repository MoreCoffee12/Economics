## Scale and append GLD to GOLDAMGBD228NLBM

Select a data range for scaling beginning with the first sample in GLD and the last sample in GOLDAMGBD228NLBM.

```r
dt.start <- index(GLD[1,0])
dt.start
```

```r
dt.end <- tail(GOLDAMGBD228NLBM,1)
dt.end <- index(dt.end[1,0])
dt.end
```

Select the data for scaling

```r
GOLDAMGBD228NLBM.scale = GOLDAMGBD228NLBM[paste(dt.start,"/",dt.end,sep="")]
GLD.Close.scale = GLD.Close[paste(dt.start,"/",dt.end,sep="")] 
```

GLD should represent the over-the-counter price of gold. Each share should be 1/10th of an ounce of gold [[1](https://www.investopedia.com/articles/investing/122515/gld-ishares-gold-trust-etf.asp#:~:text=The%20SPDR%20Gold%20Shares%20ETF%20(GLD)%20tracks%20the%20price%20of,as%20well%20as%20some%20cash.&text=Each%20share%20represents%20one%2Dtenth,of%20the%20price%20of%20gold.)]. This doesn't correspond the GOLDAMGBD228NLBM. The plots below have different values even though the GLD price has been scaled by ten.

```r
plot(GOLDAMGBD228NLBM.scale)
```

```r
plot(10*GLD.Close.scale)
```

Create a new xts object by merging these two. They have different numbers of samples (one is sourced from FRED and the other from Yahoo). For this reason, we have to remove rows with 'NA'

```r
GOLD.scale <- merge(GOLDAMGBD228NLBM.scale, GLD.Close.scale)
```

```r
GOLD.scale <- na.omit(GOLD.scale)
```

```r
plot(GOLD.scale)
```

```r
GOLD.scale
```

It doesn't look like GLD has a constant relationship to the price of gold

```r
plot(GOLD.scale[,1]/GOLD.scale[,2])
```

According to [[1](https://www.investopedia.com/articles/investing/122515/gld-ishares-gold-trust-etf.asp#:~:text=The%20SPDR%20Gold%20Shares%20ETF%20(GLD)%20tracks%20the%20price%20of,as%20well%20as%20some%20cash.&text=Each%20share%20represents%20one%2Dtenth,of%20the%20price%20of%20gold.)] the fund has an annual tracking error of 0.93%.
